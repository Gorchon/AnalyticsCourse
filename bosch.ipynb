```python
import cv2
import numpy as np
from PIL import Image

def threshold_image(image_path, output_path, threshold=128):
    # Open the image
    image = Image.open(image_path)

    # Convert the image to grayscale
    image_gray = image.convert('L')

    # Convert the grayscale image to a NumPy array
    image_array = np.array(image_gray)

    # Apply the threshold to create a binary image
    binary_image = (image_array > threshold) * 255

    # Create a new image from the binary NumPy array
    binary_image = Image.fromarray(binary_image.astype(np.uint8))

    # Save the binary image
    binary_image.save(output_path)

def count_and_draw_contours(image_path, output_path, threshold=128):
    # Threshold the image to create a binary image
    threshold_image(image_path, output_path, threshold)

    # Open the binary image using OpenCV
    binary_image = cv2.imread(output_path, cv2.IMREAD_GRAYSCALE)

    # Find contours in the binary image
    contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Create a new image to draw the contours on
    drawn_image = cv2.imread(output_path)
    drawn_image = cv2.cvtColor(drawn_image, cv2.COLOR_BGR2RGB)

    # Draw the detected contours in red color
    cv2.drawContours(drawn_image, contours, -1, (255, 0, 0), 2)

    # Convert the OpenCV image back to Pillow format
    drawn_image = Image.fromarray(drawn_image)

    # Save the image with drawn contours
    drawn_image.save(output_path)

# Specify the path to the input image file
input_image_path = "REF_23.PNG"

# Specify the path to the output image with contours
output_image_path = "REF_23_WITH_CONTOURS_2.PNG"

# Set the threshold for binarization (adjust as needed)
threshold_value = 50

# Count and draw contours on the binary image
count_and_draw_contours(input_image_path, output_image_path, threshold_value)
```

```python
import cv2
import numpy as np
from PIL import Image

def find_white_shapes(image_path, output_path):
    # Open the image
    image = Image.open(image_path)

    # Convert the image to grayscale
    image_gray = image.convert('L')

    # Convert the grayscale image to a NumPy array
    image_array = np.array(image_gray)

    # Calculate the center coordinates
    center_x, center_y = image_array.shape[1] // 2, image_array.shape[0] // 2

    # Define the width and height of the area you want to crop
    crop_width = 200  # Adjust this value as needed
    crop_height = 200  # Adjust this value as needed

    # Calculate the coordinates for cropping
    x1 = center_x - crop_width // 2
    x2 = center_x + crop_width // 2
    y1 = center_y - crop_height // 2
    y2 = center_y + crop_height // 2

    # Crop the image to the desired region
    cropped_image_array = image_array[y1:y2, x1:x2]

    # Apply Canny edge detection to the cropped image
    edges = cv2.Canny(cropped_image_array, 250, 200)

    # Find contours in the binary image
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Create a three-channel image with white lines
    white_lines = np.zeros((crop_height, crop_width, 3), dtype=np.uint8)
    white_lines[edges != 0] = (255, 255, 255)  # White color in BGR format

    # Draw the contours on the white_lines image
    cv2.drawContours(white_lines, contours, -1, (255, 0, 0), 2)

    # Save the image with white contours
    Image.fromarray(white_lines).save(output_path)

    # Get the original image for RGB comparison
    original_image = Image.open("original_image.jpg")  # Replace with the path to your original image

    # Adjust contour coordinates for the cropping
    original_contour_coords = []
    for contour in contours:
        contour_coords = []
        for point in contour:
            x, y = point[0]
            # Adjust the coordinates to account for the cropping
            x += x1
            y += y1
            original_contour_coords.append((x, y))

    # Retrieve RGB values from the original image for the adjusted coordinates
    rgb_values = []
    for x, y in original_contour_coords:
        rgb = original_image.getpixel((x, y))
        rgb_values.append(rgb)

    return contours, rgb_values

# Specify the path to the input image file
input_image_path = "processed_image.png"  # Path to the processed image with white contours

# Specify the path to the output image with white contours
output_image_path = "White_Contours.png"

# Perform edge detection and find white shapes (contours)
contours, rgb_values = find_white_shapes(input_image_path, output_image_path)

# Print the coordinates of the found contours and their corresponding RGB values
for i, contour in enumerate(contours):
    for point in contour:
        x, y = point[0]
        # Adjust the coordinates to account for the cropping
        x += x1
        y += y1
        rgb = rgb_values[i]
        print(f"Contour Coordinate: x={x}, y={y}, RGB={rgb}")
```
